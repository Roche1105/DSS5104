```python
from geopy.geocoders import Nominatim
import folium
from folium.plugins import MarkerCluster

# Calculate average price by city
city_avg_price = df.groupby('city')['price'].mean().reset_index()

# Initialize geolocator
geolocator = Nominatim(user_agent="city_geocoder")

# Function to get latitude and longitude for a city
def get_coordinates(city):
    try:
        location = geolocator.geocode(city)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except:
        return None, None

# Get coordinates for each city
city_avg_price[['latitude', 'longitude']] = city_avg_price['city'].apply(
    lambda x: pd.Series(get_coordinates(x))
)

# Drop cities with missing coordinates
city_avg_price = city_avg_price.dropna(subset=['latitude', 'longitude'])

# Create a map centered at the average latitude and longitude
map_center = [city_avg_price['latitude'].mean(), city_avg_price['longitude'].mean()]
city_map = folium.Map(location=map_center, zoom_start=10)

# Add markers for each city
marker_cluster = MarkerCluster().add_to(city_map)
for _, row in city_avg_price.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"City: {row['city']}<br>Average Price: ${row['price']:.2f}",
    ).add_to(marker_cluster)

# Display the map
city_map
```

```python
from folium import Map
from folium.plugins import HeatMap
# Prepare data for the heatmap
heatmap_data = city_avg_price[['latitude', 'longitude', 'price']].values.tolist()

# Create a map centered at the average latitude and longitude
map_center = [city_avg_price['latitude'].mean(), city_avg_price['longitude'].mean()]
heatmap_map = Map(location=map_center, zoom_start=8)

# Add heatmap layer
HeatMap(heatmap_data, radius=15).add_to(heatmap_map)

# Save the map to an HTML file
heatmap_map.save("city_price_heatmap.html")

# Display the map (if running in a Jupyter Notebook)
heatmap_map
```
